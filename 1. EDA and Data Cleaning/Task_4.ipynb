```python
import pandas as pd
import numpy as np
import glob
import os

info_path = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /Other/Match_Info.csv'
match_info_df = pd.read_csv(info_path, index_col=0)

team_path = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /Other/Team_Info.csv'
team_info_df = pd.read_csv(team_path, index_col=0)
```

```python
# rename the index

match_info_df = match_info_df.rename_axis('Link').reset_index()
```

```python
def format_ref_names(referee):
    referee = referee.replace("\r\n", "")
    referee = referee.replace("Referee: ", "")
    return referee

def format_link(link):
    link = "https://www.besoccer.com" +link 
    return link
```

```python
# We'll iterate over the rows of the dataframe, and reassign the row to the standardised variant
for index, row in match_info_df.iterrows():
    
        
    ## Call our standardisation function on the phone number for the current loop
    referee = row["Referee"] 
    link = row["Link"]

    referee_formayted = format_ref_names(referee)
    link_formatted = format_link(link)
    #print(phone_number_stand)
    match_info_df.at[index, "Referee"] = referee_formayted
    match_info_df.at[index, "Link"] = link_formatted

    
    

match_info_df
```

```python
import joblib

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder



path_to_predict = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /Scraped Datasets/to_predict_prepared_fo_ML_df.csv'
path_premiership_data = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /Cleaned Datasets/ML_split/classification_premiership_decade_3_dataset.csv'

clean_epl_d3_df = pd.read_csv(path_premiership_data)
clean_epl_d3_df.rename(columns={"ELO_home" : "ELO_Home", "ELO_away" : "ELO_Away"}, inplace=True)


path_rfc_model = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /match-outcome-predictor/4. Model Training/Models/RandomForestClassifier.joblib'
path_knn_model = r'/Users/tom/Documents/Coding/AiCore/Projects/4. Football Match Outcome Predictor /match-outcome-predictor/4. Model Training/Models/KNeighbours.joblib'

rfc_model = joblib.load(path_rfc_model)
kn_model = joblib.load(path_knn_model)


to_predict_prepared_fo_ML_df = pd.read_csv(path_to_predict)

to_predict_prepared_fo_ML_df
```

```python
# Compute Training data normaliser

X_modelling_set = clean_epl_d3_df.drop(columns=["Home_Result", 
"away_team_current_goal_drought",
"home_team_current_goal_drought",
"home_team_current_loss_streak",
"away_team_current_loss_streak",
"home_team_current_win_streak", 
"away_team_current_win_streak", 
"home_team_total_wins_so_far", 
"away_team_total_wins_so_far", 
"home_team_total_goals_conceeded_so_far",
"away_team_total_goals_conceeded_so_far"], axis=1)
y_modelling_set = (clean_epl_d3_df["Home_Result"])

# Encode the labels of the targets

le = LabelEncoder()
# Fit the encoder to the targets column
le.fit(y_modelling_set)
#list the classes of the targets
#list(le.classes_)
# transform the targets column (inverse_transform to change it back!)
y_encoded = le.transform(y_modelling_set)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_modelling_set, y_encoded, test_size = 0.3)

min_max_scaler = MinMaxScaler()

# Features normalised by the MinMax approach

min_max_scaler.fit_transform(X_train)
```

```python
# Using the tuned RFC model

# Drop team names and non required features
condensed_fixtures = to_predict_prepared_fo_ML_df[["Home_Team","Away_Team"]]
features = to_predict_prepared_fo_ML_df.drop(columns=["Home_Team", 
"Away_Team", 
"away_team_current_goal_drought",
"home_team_current_goal_drought",
"home_team_current_loss_streak",
"away_team_current_loss_streak",
"home_team_current_win_streak", 
"away_team_current_win_streak", 
"home_team_total_wins_so_far", 
"away_team_total_wins_so_far", 
"home_team_total_goals_conceeded_so_far",
"away_team_total_goals_conceeded_so_far"], axis=1)

# Normalise data using modelling set min max normaliser

features_min_max = min_max_scaler.transform(features)

# get the predictions

rfc_predicitions = rfc_model.predict(features_min_max)

# decode the predicitions

labels = {'draw' : 0, 'loss' : 1, 'win' : 2}
rfc_unencoded_predicitions = ['draw' if x == 0 else 'loss' if x == 1 else 'win' for x in rfc_predicitions ]
condensed_fixtures["RFC Predictions"] = rfc_unencoded_predicitions
predicted_table = to_predict_prepared_fo_ML_df.copy()
predicted_table["RFC Predictions"] = rfc_unencoded_predicitions

# organise into a more reable format

predicted_table.drop(columns=["away_team_current_goal_drought",
"home_team_current_goal_drought",
"home_team_current_loss_streak",
"away_team_current_loss_streak",
"home_team_current_win_streak", 
"away_team_current_win_streak", 
"home_team_total_wins_so_far", 
"away_team_total_wins_so_far", 
"home_team_total_goals_conceeded_so_far",
"away_team_total_goals_conceeded_so_far"], inplace=True)



```

```python
#Using the tuned Kneighbours model

# get the predictions

kn_predicitions = kn_model.predict(features_min_max)

# decode the predicitions

labels = {'draw' : 0, 'loss' : 1, 'win' : 2}
kn_unencoded_predicitions = ['draw' if x == 0 else 'loss' if x == 1 else 'win' for x in kn_predicitions ]
condensed_fixtures["KN Predictions"] = kn_unencoded_predicitions
predicted_table["KN Predictions"] = kn_unencoded_predicitions
```
